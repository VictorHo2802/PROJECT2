## 1. THE POOLED OLS MODEL

```python

# Thêm một hằng số nhỏ vào các giá trị trước khi tính log để tránh lỗi chia cho 0
df['log_GPA'] = np.log(df['GPA'].replace(0, np.nan))  # Thay thế giá trị 0 bằng NaN trước khi tính log
df['log_ThoiGianHoc'] = np.log(df['ThoiGianHoc'].replace(0, np.nan))
df['log_ThoiGianNgoaiKhoa'] = np.log(df['ThoiGianNgoaiKhoa'].replace(0, np.nan))
df['log_ThoiGianNgu'] = np.log(df['ThoiGianNgu'].replace(0, np.nan))
df['log_ThoiGianHoatDongTheChat'] = np.log(df['ThoiGianHoatDongTheChat'].replace(0, np.nan))


# Loại bỏ các dòng chứa NaN (do giá trị 0 đã được thay thế bằng NaN)
df = df.dropna(subset=['log_GPA', 'log_ThoiGianHoc', 'log_ThoiGianNgoaiKhoa', 'log_ThoiGianNgu', 'log_ThoiGianHoatDongTheChat'])

# Xây dựng mô hình hồi quy tuyến tính với log của GPA làm biến phụ thuộc
model = smf.ols(
    formula="log_GPA ~ log_ThoiGianHoc + log_ThoiGianNgoaiKhoa + log_ThoiGianNgu + log_ThoiGianHoatDongTheChat",
    data=df
)

# Tính toán và in kết quả
results = model.fit()
print(results.summary())


```

```python
df = df.set_index(['Ma_SV , ThoiGianHoc'])

# Biến phụ thuộc: log của GPA
y = df['GPA']

# Biến độc lập: log của các yếu tố ảnh hưởng
X = df[['ThoiGianHoc', 'ThoiGianNgoaiKhoa', 'ThoiGianNgu', 'ThoiGianHoatDongTheChat']]

# Thêm hằng số vào biến độc lập
X = X.assign(constant=1)

# Mô hình Fixed Effects (FE) với entity_effects=True
model = PanelOLS(y,X, entity_effects=True)

# Ước lượng mô hình
fe_results = model.fit()

# In kết quả
print(fe_results.summary)
```
