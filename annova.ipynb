```python
import pandas as pd
from scipy.stats import f_oneway
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

df = pd.read_excel("/Users/macbook/Desktop/TAXI_PRICE_EXTENSION/DATA/DATA HỖN HỢP/annova.xlsx", sheet_name="Sheet1")

grouped = df.groupby("group")["score"].apply(list)

f_stat, p_val = f_oneway(*grouped)
print(f"=== SciPy one-way ANOVA ===")
print(f"F = {f_stat:.4f}, p-value = {p_val:.4f}\n")

model = ols("score ~ C(group)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("=== statsmodels ANOVA table ===")
print(anova_table)
```

```python
tukey = pairwise_tukeyhsd(endog=df["score"],
                          groups=df["group"],
                          alpha=0.05)
print(tukey.summary())
```

```python
groups = [grp["score"].values for name, grp in df.groupby("group")]
f_stat, p_anova = f_oneway(*groups)
print(f"One-way ANOVA: F = {f_stat:.4f}, p-value = {p_anova:.4f}")

model = ols("score ~ C(group)", data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("\nANOVA table (statsmodels):")
print(anova_table)

tukey = pairwise_tukeyhsd(endog=df["score"],
                          groups=df["group"],
                          alpha=0.05)

print("\nTukey:")
print(tukey.summary())

tukey = pairwise_tukeyhsd(endog=df["score"],
                          groups=df["group"],
                          alpha=0.05)

data = tukey._results_table.data

df_tukey = pd.DataFrame(data[1:], columns=data[0])

print(df_tukey[["group1", "group2", "p-adj"]])

```
