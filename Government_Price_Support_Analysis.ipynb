
# Government Price Support and Loan Assistance Policy

This notebook illustrates how government intervention through price support and loan assistance to firms impacts market outcomes. Specifically, it shows how the government sets a support price above equilibrium and commits to buying the excess supply, often using loan-backed purchases.

## Key Concepts Illustrated:
- **Consumer Surplus**: Area under the demand curve and above the market price.
- **Producer Surplus**: Area above the supply curve and below the market price.
- **Government Purchases**: The area representing excess supply the government agrees to buy.
- **Deadweight Loss**: Efficiency loss due to misallocation of resources.


```python

import matplotlib.pyplot as plt
import numpy as np

# Định nghĩa các hàm cung và cầu
def demand(q): return 20 - q
def supply(q): return 2 + 0.8*q

q = np.linspace(0, 10, 100)
p_d = demand(q)
p_s = supply(q)

# Giá hỗ trợ
p_support = 10

# Tính toán giao điểm (giá và sản lượng cân bằng)
q_eq = 6
p_eq = demand(q_eq)

# Sản lượng theo cung tại giá hỗ trợ (chính phủ mua phần dư)
q_supply = (p_support - 2) / 0.8
q_demand = 20 - p_support

fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(q, p_d, label='Demand (D)', color='blue')
plt.plot(q, p_s, label='Supply (S)', color='red')
plt.axhline(p_support, linestyle='--', color='purple', label='Support Price ($10)')

# Vùng thặng dư tiêu dùng
plt.fill_between(q, p_support, p_d, where=(q <= q_demand), color='skyblue', alpha=0.5, label='Consumer Surplus')
# Vùng thặng dư sản xuất
plt.fill_between(q, p_s, p_support, where=(q <= q_supply), color='lightgreen', alpha=0.5, label='Producer Surplus')
# Vùng chính phủ mua vào
plt.fill_betweenx([0, p_support], q_demand, q_supply, color='khaki', alpha=0.5, label='Government Purchases (Loan-supported)')
# Deadweight loss
plt.fill_between(q, p_d, p_s, where=((q > q_demand) & (q < q_supply)), color='orange', alpha=0.5, label='Deadweight Loss')

plt.xlabel('Quantity (million tons/year)')
plt.ylabel('Price ($/ton)')
plt.title('Government Price Support and Loan Assistance to Firms')
plt.legend()
plt.grid(True)
plt.show()
```
